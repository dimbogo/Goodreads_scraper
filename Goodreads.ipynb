```python
import requests
from bs4 import BeautifulSoup

page_num = 1
book_dict = {}
for i in range(100):
  url = f'https://www.goodreads.com/list/show/1.Best_Books_Ever?page={page_num}'
  response = requests.get(url.format(page_num))
  soup = BeautifulSoup(response.text, 'html.parser')

  # Scrape data from the page here
  for book in soup.find_all('tr', itemtype='http://schema.org/Book'):
    title = book.find('a', class_='bookTitle').text.strip()
    rating = book.find('span', class_='minirating').text.strip().split()
    votes = book.find('span', class_='minirating').text.strip().split()
    if rating[0] == 'really' or rating[0] == 'it':
      rating = rating[3]
      votes = votes[7]
    else:
      rating = rating[0]
      votes = votes[4]
    book_dict[title] = [rating, votes]

  page_num += 1

sorted_book_dict = sorted(book_dict.items(), key=lambda x:x[1][0], reverse = True)
```

```python
important_books = [k for k in sorted_book_dict if int(k[1][1].replace(',','')) > 9999]
```
